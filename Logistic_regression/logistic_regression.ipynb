```python
from sklearn.datasets import fetch_openml
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist.data[0:1]
```

```python
mnist.target[:]
```

```python
mnist.target = [int(i) for i in mnist.target]
```

```python
mnist.target
```

```python
plt.figure(figsize = (20,4))
for index, (image, label) in enumerate(zip(mnist.data[0:5],mnist.target[0:5])):
    plt.subplot(1, 5, index+1)
    plt.imshow(np.reshape(image,(28, 28)), cmap = 'Greys')
    plt.title('Label%i'%label, fontsize = 15)
plt.show()
```

```python
X = mnist.data
Y = mnist.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10, random_state = 0)
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver = 'lbfgs', max_iter = 1e1)
logreg.fit(X_train, Y_train)
```

```python
logreg.score(X_test, Y_test)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
predict = logreg.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, predict)
```
