```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edges_from([('A', 'E'), ('B', 'A'), ('C', 'A'), ('D', 'B'), ('D', 'C'), ('E', 'C'), ('E', 'D')])
```

```python
plt.figure(figsize =(10, 10))
nx.draw_networkx(G, with_labels = True)
```

```python
p = nx.pagerank(G, alpha = 0.85)
print(p)
```
